Now an important question. 
Can high GRE scores cover up for low CGPA and vice-versa? 
Let's check it out!

So for better understanding, we will be creating bins for CGPA in following manner:
1. Category 1 for High CGPA
2. Category 2 for Average CGPA
3. Category 3 for low CGPA

```python
dataset['CGPA_cat']=0
```

```python
temp = [dataset]
```

```python
for data in temp:
    data.loc[dataset['CGPA']<=8,'CGPA_cat']=3,
    data.loc[(dataset['CGPA']>8) & (dataset['CGPA']<=9),'CGPA_cat']=2,
    data.loc[(dataset['CGPA']>9) & (dataset['CGPA']<=10),'CGPA_cat']=1
    
```

As shown, High GRE scores cannot cover up for high cgpa as the difference between aspirants with same GRE scores but different CGPA is clearly visible.

# Data Pre-processing

Here, we will check for:
* missing values
* covert categorical variables(since we have all variables in number for so no need)
* do some feature engineering(if necessary).

Let's check for missing values

```python
dataset.isnull().sum()
```

Now we will split dataset into train and target variables

```python
target=dataset['Admit_rate']
```

```python
drop=['Serial No.', 'Admit_rate','CGPA_cat']
train = dataset.drop(drop, axis=1)
```

Now it is time to build model.

# Machine Learning Models

Now we will split the dataset in the ratio of 75:25 for train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, target, test_size = 0.25, random_state = 0)
```

* XGBoost Regressor

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

model = XGBRegressor(n_estimators=1000,learning_rate=0.05,n_jobs=-1)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
def rmse(y_test,y_pred):
      return np.sqrt(mean_squared_error(y_test,y_pred))

y_pred = model.predict(X_test)
print("XGBoost score on testing set: ", rmse(y_test, y_pred))
```

```python
XGB_acc = round(model.score(X_test, y_test)*100, 2)
print("Accuracy of XGB Model={}".format(XGB_acc))
```

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.title("Chance of Admit Test and predicted data")
plt.legend()
plt.show()
```

Feature Scaling the data for linear models

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
  
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

* Lasso Regression

```python
from sklearn.linear_model import Lasso
best_alpha = 0.0099

regr = Lasso(alpha=best_alpha, max_iter=50000)
regr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
def rmse(y_test,y_pred):
      return np.sqrt(mean_squared_error(y_test,y_pred))
y_pred = regr.predict(X_test)
print("Lasso score on testing set: ", rmse(y_test, y_pred))
```

```python
lasso_acc = round(regr.score(X_test, y_test)*100, 2)
print("Accuracy of Lasso Model={}".format(lasso_acc))
```

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.title("Chance of Admit Test and predicted data")
plt.legend()
plt.show()
```

* Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
def rmse(y_test,y_pred):
      return np.sqrt(mean_squared_error(y_test,y_pred))

y_pred = lr.predict(X_test)
print("Linear Regressor score on testing set: ", rmse(y_test, y_pred))
```

```python
lr_acc = round(lr.score(X_test, y_test)*100, 2)
print("Accuracy of Linear Regression Model={}".format(lr_acc))
```

```python
x_ax = range(len(y_test))
plt.plot(x_ax, y_test, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.title("Chance of Admit Test and predicted data")
plt.legend()
plt.show()
```

So, XGBoost has been giving the best results when compared to other 2 models.
